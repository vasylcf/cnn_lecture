```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('samples/city.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(12,7))
plt.imshow(gray, cmap='gray')
```

```python
# construct the Sobel x-axis kernel
sobelX = np.array((
	[-1, 0, 1],
	[-2, 0, 2],
	[-1, 0, 1]), dtype="int")
 
# construct the Sobel y-axis kernel
sobelY = np.array((
	[-1, -2, -1],
	[0, 0, 0],
	[1, 2, 1]), dtype="int")
```

```python
plt.figure(figsize=(12,7))
plt.imshow(cv2.filter2D(gray, ddepth=-1, kernel=sobelX), cmap='gray')
```

```python
plt.figure(figsize=(12,7))
plt.imshow(cv2.filter2D(gray, ddepth=-1, kernel=sobelY), cmap='gray')
```
